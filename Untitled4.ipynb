```python
def get_data_and_param(path, numepoch = 1, batch_size = 8):
    path1 = path
    files = os.listdir(os.path.join(path1, 'blurred'))
    l1= list()
    d1 = list()
    random.shuffle(files)
    count = 0
    for i in files:
            try:
              data = np.load(os.path.join(path1, 'blurred', i)).astype(np.float32)
              label = np.load(os.path.join(path1, 'orig', i)).astype(np.float32)
              data = (data - np.amin(data)) / (np.amax(data) - np.amin(data))
            except Exception:
              continue
            #print(data.shape, label.shape)
            for q in range(2):
                j = random.randint(0, data.shape[0] - 1)
                if math.isnan(np.amax(data[j])) or math.isnan(np.amax(label[j])):
                  q = q - 1
                  continue
                count += 1
                if (count > 1024 and path == patht) or (count > 500 and path == pathv):
                  return
                d1 = (data[j].transpose(1, 2, 0))
                l1 = (label[j].transpose(1, 2, 0))
                yield (np.array(d1), np.array(l1))
def get_data_and_param_t(numepoch = 1, batch_size = 8):
  for i in get_data_and_param(patht):
    yield i

def get_data_and_param_v(numepoch = 1, batch_size = 8):
  count = 0
  for i in get_data_and_param(pathv):
    yield i
```

```python

data_train = np.empty((1024, 128, 128, 10))
label_train = np.empty((1024, 128, 128, 10))
j = 0
for i in get_data_and_param_t():
  clear_output(wait = True)
  print(j)
  data_train[j] = i[0]
  label_train[j] = i[1]
  j += 1
```

```python
train_dataset = tf.data.Dataset.from_tensor_slices((data_train, label_train)).batch(8)
```

```python
def mssim(X, Y):
  return tf.image.ssim_multiscale(X, Y, 1, filter_size = 6)

def lssim(X, Y):
  return 1 - tf.image.ssim_multiscale(X, Y, 1, filter_size = 6)

def psnr1(X, Y):
  return psnr(X, Y, 1)

dependencies = {
    'lssim':lssim,
    'mssim' : mssim,
    'psnr1' : psnr1
}
```
